### What is Machine Learning Model
**Machine Learning model is a program that maps inputs(Feautures) to outputs(Labels/predictions) bassed on learning patterns from the data**

### Linear Regression Model
**Linear Regression is a linear relationship between the independent and dependent variables. 
This means that the change in the dependent variable is proportional to the 
change in the independent variables. For example predicting the price of a house based on its size.**

**if its Linear, we need to check no. of Feautures or no. of independent variable
if we have 1 independent variable or feature, we can use 'Linear Regression'
A simple Linear Regression Equation is:**
    
    * y = mX + b

     m is slope of the line, b is the y-intercept
     y is the output / Dependent variable
     x is the input / Independent Variable

### What is Estimator?
Estimator is an object in machine learning that can be trained on data and make prediction

* fit(X,y): Learns from data
* predict(X): Make predictions
* score(X,y): Evaluate performance

Most models in **scikit-learn**, **TensorFlow/Keras**, and **PyTorch** are estimator

## Machine Learning Project - 6. Tune and Improve - ML model; Hyperparameters

* Improve Model Predictions through experiments (Hyperparameter Tuning)
* Manual hyperparameter tuning
* Hyperparameter tuning using: RandomizedSearchCV and GridSearchCV

```python
# 1. Setup random seed - to have the same results, me and you
import numpy as np
np.random.seed(42)
```

```python
# 2. Load the data using 'pandas'
import pandas as pd
house = pd.read_csv(r"C:\Users\Admin\Datasets\train.csv")
```

```python
# 3. Split the data into 'features' and 'labels'
X = house.drop('Saleprice', axis=1)
y = house["Saleprice"]
```

```python
# 4. Split the Data into 'training set' and 'test set'
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
# 5. Create Machine Learning Model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()

# model = RandomForestRegressor(
#    n_estimators=50,
#    max_depth=10,
#    min_samples_split=5)
```

```python
# 6. a) Traing the Model
model.fit(X_train, y_train)
```

```python
# 6. b) Make predictions using ML model
y_preds = model.predict(X_test)
```

```python
# 7. Evaluate the Model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate metrics
mae = mean_absolute_error(y_test, y_preds)
mse = mean_squared_error(y_test, y_preds)
r2  = r2_score(y_test, y_preds)

# print the formatted results
print("Mean Absolute Error : ", mae)
print("Mean Squared Error  : ",mse)
print("R-Squared Error     : ",r2)
```

```python
# 8. a) Tune and Improve Machine Learning Model
model.get_params()
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Hyperparameter tuning using RandomizedSearchCV
param_dist = {
    'n_estimators' : [50, 100, 200],     # Number of trees in the forest
    'max_depth' : [None, 10, 20, 30],    # Maximum depth of the tree
    'min_samples_split' : [2, 5, 10]     # Minimum number of samples required to split an internal node
}

random_search_cv = RandomizedSearchCV(
    model,
    param_distributions=param_dist,
    n_iter=5, cv=5)
```

```python
random_search_cv.fit(X_train, y_train)
```

```python
# 8. b) Best Model

# Best Hyperparameters and scores
print("Best Hyperparameters : ",random_search_cv.best_params_)
print("Best Cross-Validation Scores : ", random_search_cv.best_score_)
```

```python
# 8. c) Evaluate Results on the Test Set
# For the optimized model using RandomiRandomizedSearchCV
best_model = random_search_cv.best_estimator_
print("Test set results - accuracy : ", best_model.score(X_test, y_test))
```
